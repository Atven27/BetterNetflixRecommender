```python
#Reading and accumulating data, formatting
import pandas as pd
import numpy as np
import os
import ast
import re

# genre: convert string representation of list to actual list
def convert(s):
    s=re.sub("[^A-Za-z,]","",string)
    return list(s.split(","))

def convert2(s):
    l = ast.literal_eval(s)
    st = ''
    for item in l:
        st = item['name'] + ' ' + st
    return st.lower()

def convert3(s):
    l = [s]
    return l

df1 = pd.read_csv("netflix_titles.csv", sep=',', encoding='latin-1')
df2 = pd.read_csv("netflix_originals.csv", sep=',', encoding='latin-1')

df1 = df1[['type', 'title', 'director', 'cast', 'description']]
df1['title'] = df1['title'].str.upper()
df1['type'] = df1['type'].str.upper()
df1 = df1.rename(columns={'type':'type1'})
df1 = df1.rename(columns={'description':'description1'})
df1['cast'] = [x.split(' ') for x in df1['cast'].astype(str)]
#df1.columns

df2 = df2[['Title', 'Genre', 'IMDB Score']]
df2 = df2.rename(columns={'Title':'title'})
df2 = df2.rename(columns={'Genre':'genre1'})
df2 = df2.rename(columns={'IMDB Score':'imdb_score1'})
df2['title'] = df2['title'].str.upper()
df2['genre1'] = df2['genre1'].str.lower()
#df2.columns

df3 = df1.merge(df2, on  = "title",how="outer") #, indicator=True
df3['source'] = 'Netflix'
#print(len(df3.index))
#df3.head(2)

df_titles = pd.read_csv("titles.csv", sep=',', encoding='latin-1')
df_titles = df_titles[['type', 'title', 'description', 'genres', 'imdb_id', 'imdb_score', 'imdb_votes', 'tmdb_popularity','tmdb_score']]
df_titles['title'] = df_titles['title'].str.upper()
df_titles['type'] = df_titles['type'].str.upper()
df_titles = df_titles.rename(columns={'imdb_score':'imdb_score2'})
df_titles = df_titles.rename(columns={'genres':'genre2'})
df_titles = df_titles.rename(columns={'type':'type2'})
df_titles = df_titles.rename(columns={'description':'description2'})
df_titles['genre2'] = df_titles['genre2'].apply(convert)
#df_titles.columns
#df_titles.head(1)

df4 = df3.merge(df_titles, on  = "title",how="left") #, indicator=True
#print(len(df4.index))
#df4.head(2)

df_10ktitles = pd.read_csv("10k_titles.csv", sep=',', encoding='latin-1')
df_10ktitles = df_10ktitles[['Genres', 'title', 'Keywords', 'overview', 'popularity', 'vote_average', 'vote_count']]
df_10ktitles['title'] = df_10ktitles['title'].str.upper()
df_10ktitles = df_10ktitles.rename(columns={'Genres':'genre3'})
df_10ktitles = df_10ktitles.rename(columns={'overview':'description3'})
df_10ktitles = df_10ktitles.rename(columns={'vote_average':'vote_average1'})
df_10ktitles = df_10ktitles.rename(columns={'vote_count':'vote_count1'})
df_10ktitles = df_10ktitles.rename(columns={'Keywords':'keywords'})
df_10ktitles['genre3'] = df_10ktitles['genre3'].apply(convert2)
df_10ktitles['keywords'] = df_10ktitles['keywords'].apply(convert2)
df_10ktitles['keywords'] = df_10ktitles['keywords'].apply(convert3)

df_10ktitles.head(1)

df5 = df4.merge(df_10ktitles, on  = "title",how="left") #, indicator=True
#print(len(df5.index))
#df5.head(1)

df_mm = pd.read_csv("movies_metadata.csv", sep=',', encoding='latin-1', low_memory=False)
df_mm = df_mm[['overview', 'popularity', 'title', 'genres', 'imdb_id', 'vote_average', 'vote_count']]
df_mm['title'] = df_mm['title'].str.upper()
df_mm = df_mm.rename(columns={'genres':'genre4'})
df_mm = df_mm.rename(columns={'overview':'description4'})
df_mm = df_mm.rename(columns={'imdb_id':'imdb_id2'})
df_mm = df_mm.rename(columns={'vote_average':'vote_average2'})
df_mm = df_mm.rename(columns={'vote_count':'vote_count2'})
df_mm = df_mm.rename(columns={'popularity':'popularity2'})
df_mm['genre4'] = df_mm['genre4'].apply(convert2)
df_mm.head(2)

df6 = df5.merge(df_mm, on  = "title",how="left") #, indicator=True
df6[['imdb_score1','imdb_score2']] = df6[['imdb_score1','imdb_score2']].fillna(value=0)
df6[['tmdb_popularity','popularity','popularity2']] = df6[['tmdb_popularity','popularity','popularity2']].fillna(value=0)
df6[['vote_count1','vote_count2']] = df6[['vote_count1','vote_count2']].fillna(value=0)
df6[['vote_average1','vote_average2']] = df6[['vote_average1','vote_average2']].fillna(value=0)
df6[['tmdb_score']] = df6[['tmdb_score']].fillna(value=0)

# replace NANs to blank string
df6 = df6.fillna('')
#print(len(df6.index))
#df6.head(1)

#type = type1+type2
#genre = genre1+genre2+genre3+genre4
#imdb_score = max(imdb_score1,imdb_score2,tmdb_score)
#imdb_id = imdb_id+':'+imdb_id2
#popularity = max(tmdb_popularity,popularity,popularity2)
#vote_count = max(vote_count1,vote_count2)
#vote_average = max(vote_average1,vote_average2)
#description = description1+description2+description3+description4

df7 = df6
df7['type']=df7['type1']+':'+df7['type2']
#df7['genre']=df7['genre1']+' '+df7['genre2']+' '+df7['genre3']+' '+df7['genre4']

df7['genre'] = np.where(df7['genre1'].str.len() > 0,df7['genre1'],'')
df7['genre'] = np.where(df7['genre2'].str.len() > 0,df7['genre']+' '+df7['genre2'],df7['genre'])
df7['genre'] = np.where(df7['genre3'].str.len() > 0,df7['genre']+' '+df7['genre3'],df7['genre'])
df7['genre'] = np.where(df7['genre4'].str.len() > 0,df7['genre']+' '+df7['genre4'],df7['genre'])

df7['imdb_score1'] = df7['imdb_score1'].astype(float)
df7['imdb_score2'] = df7['imdb_score2'].astype(float)
df7['tmdb_score'] = df7['tmdb_score'].astype(float)
df7['imdb_score'] = df7[['imdb_score1', 'imdb_score2','tmdb_score']].values.max(1)

df7['imdb_id']=df7['imdb_id']+':'+df7['imdb_id2']
df7['tmdb_popularity'] = df7['tmdb_popularity'].astype(float)
df7['popularity'] = df7['popularity'].astype(float)
#df7['popularity2'] = df7['popularity2'].astype(float)
df7['popularity_score'] = df7[['tmdb_popularity', 'popularity']].values.max(1)

df7['vote_count1'] = df7['vote_count1'].astype(float)
df7['vote_count2'] = df7['vote_count2'].astype(float)
df7['vote_count'] = df7[['vote_count1', 'vote_count2']].values.max(1)

df7['vote_average1'] = df7['vote_average1'].astype(float)
df7['vote_average2'] = df7['vote_average2'].astype(float)
df7['vote_average'] = df7[['vote_average1', 'vote_average2']].values.max(1)

#df7.loc[df7['description1'].isnull(), 'description'] = df7['description2']
#df7.loc[df7['description2'].isnull(), 'description'] = df7['description3']
#df7.loc[df7['description3'].isnull(), 'description'] = df7['description4']
df7['description1'] = df7['description1'].astype(str)
df7['description2'] = df7['description2'].astype(str)
df7['description3'] = df7['description3'].astype(str)
df7['description4'] = df7['description4'].astype(str)

df7['description'] = np.where(df7['description1'].str.len() > 0,df7['description1'],'')
df7['description'] = np.where(df7['description'].str.len() == 0,df7['description2'],df7['description'])
df7['description'] = np.where(df7['description'].str.len() == 0,df7['description3'],df7['description'])
df7['description'] = np.where(df7['description'].str.len() == 0,df7['description4'],df7['description'])
#df7['description'] = np.where(df7['description3'].str.len() > 0 and df7['description'].str.len() == 0,df7['description3'],'')
#df7['description'] = np.where(df7['description4'].str.len() > 0 and df7['description'].str.len() == 0,df7['description4'],'')

df7 = df7.drop(columns=['type1','type2','genre1','genre2','genre3','genre4','imdb_score1','imdb_score2','tmdb_score','imdb_id2'])
df7 = df7.drop(columns=['tmdb_popularity', 'popularity','popularity2','vote_count1','vote_count2','vote_average1','vote_average2'])
df7 = df7.drop(columns=['description1', 'description2','description3','description4'])

#df7 = df7.drop_duplicates(subset = ['title'],keep = 'last').reset_index(drop = True)
df7.insert(0, 'MyID', range(1, 1 + len(df7)))

#df7.head(10)

df8 = df7.loc[(df7['source'] == 'Netflix')] #smaller dataset to accomodate matrix
df8['description']=df8['description'].apply(lambda x: x.split())
df8['genre'] = [x.split(' ') for x in df8['genre']]

#--------------------Data Accumulation ends here------------------------

#Scoring
C = df7["vote_average"].mean()
m = df7["vote_count"].quantile(0.9)
q_movies = df8.copy().loc[df8["vote_count"] >= m]
q_movies.shape

def weighted_rating(x, m=m, C=C):
    v = x["vote_count"]
    R = x["vote_average"]
    return (v/(v+m) * R) + (m/(m+v) * C)

q_movies["imdb_score2"] = q_movies.apply(weighted_rating, axis = 1)
q_movies = q_movies.sort_values("imdb_score2", ascending = False)
q_movies['demog_score'] = q_movies[['imdb_score2', 'imdb_score']].values.max(1)

q_movies = q_movies[['title','demog_score']]
q_movies.head(1)

# Bring in the score and create tag field
df9 = pd.merge(df8,q_movies,how="left",on=['title']) #, indicator=True
#df9 = df8.join(q_movies.set_index(['title']),how="left",on=['title']) #, indicator=True
#df9.dropna(inplace=True)

cols = ['genre', 'description']
#cols = ['genre', 'description','cast','keywords']
df9['tags'] = [[e for e in row if e==e] for row in df9[cols].values.tolist()] # tags field is created here
#df9['tags'] = df9['genre'] + df9['description']

#df9 = df9.drop_duplicates(subset = ['MyID', 'title'],keep = 'last').reset_index(drop = True)
#df9.head(5)

#lemmitization and porter steammer
partial_df = df9[["MyID", "title", "tags"]]
#partial_df['tags'] = partial_df['tags'].apply(convert)
partial_df['listtags'] = [','.join(map(str, l)) for l in partial_df['tags']]
partial_df['listtags'] = partial_df['listtags'].apply(lambda x: " ".join(x))
partial_df['listtags'] = partial_df['listtags'].apply(lambda x: x.lower())
#partial_df.head(2)

import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords

nltk.download('wordnet')
nltk.download('omw-1.4')

wnl = WordNetLemmatizer()

def lemmatize_convert(text):
    words = nltk.word_tokenize(text)
    words = [lemmatizer.lemmatize(word) for word in words if word not in set(stopwords.words('english'))]
    return ' '.join(words) 

partial_df['listtags'] = partial_df['listtags'].apply(lemmatize_convert)
partial_df.head(10)

df10 = df9.merge(partial_df[['MyID','title','listtags']], on  = ['MyID','title'],how="inner") #, indicator=True
#df10.head(2)

#df10 is the final dataset after lemmitization and porter steammer
#save the final dataset into a CSV for further processing

op_csv_name = 'prep_movie_list.csv'
op_path = os.getcwd() + "/" + op_csv_name
df10.to_csv(op_path)
```
